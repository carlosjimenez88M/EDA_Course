```python
## Librerias necesarias
import pandas as pd 
import matplotlib.pyplot as plt


test = pd.read_csv('C:/Users/EDUARDO/EDA_Course/test.csv')
```

```python
print('Validando la cantidad de datos')
print('='*32)
print(test.shape)
print('='*32)
```

```python
print('Validando la estructura de los datos')
print('='*32)
print(test.info())
print('='*32)
```

```python
# Drop some unimportant columns in the dataset.
test.drop(['Name','PassengerId', 'Pclass', 'Sex', 'SibSp', 'Ticket','Cabin','Parch'],axis=1, inplace=True)
```

```python
print('Validando el comportamienton general de los datos')
print('='*64)
print(test.describe())
print('='*64)
```

```python
import seaborn as sns
sns.jointplot(x='Age', 
              y='Fare', 
              data=test, 
              color ='orange', 
              kind ='scatter', 
              hue='Embarked',
              size = 10.0)
```

```python
sns.pairplot(test, hue='Embarked')  
plt.show()
```

```python
g = sns.FacetGrid(test, col='Embarked',col_wrap=2,sharex=True,sharey=True)
g.map(sns.scatterplot,'Age','Fare')
```
