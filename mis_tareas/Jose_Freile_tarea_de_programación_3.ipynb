## Tarea #1 

Con base al dataframe de pokemon (El link se encuentra en la intro de la clase), va a crear un for loop con if cuando el poder promedio de los pokemones es superior se debe imprimir el mensaje 'Este pokemon' + nombre del pokemon + 'tiene mayor poder promedio ya que su fuerza es de ' + Fuerza . En el caso contrario 'Este pokemon' + nombre del pokemon + 'tiene menor poder promedio ya que su fuerza es de ' + Fuerza

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pokemon = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')
```

```python
pokemon.head()
```

```python
promedio =  pokemon['Total'].mean()
print('--------------------------------- Poder Promedio: ' + format(promedio))
for i in pokemon.index:
    if pokemon.iloc[i]['Total'] > promedio:
        print ('Este pokemon: ' + pokemon.iloc[i]['Name'] + ' tiene MAYOR poder promedio ya que su fuerza es de ' + format(pokemon.iloc[i]['Total']))
    else :
        print ('Este pokemon: ' + pokemon.iloc[i]['Name'] + ' tiene MENOR poder promedio ya que su fuerza es de ' + format(pokemon.iloc[i]['Total']))
   
```
