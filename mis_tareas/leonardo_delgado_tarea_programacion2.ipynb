Con base al dataframe de pokemon (El link se encuentra en la intro de la clase), va a crear un for loop con if cuando el poder promedio de los pokemones es superior se debe imprimir el mensaje 'Este pokemon' + nombre del pokemon + 'tiene mayor poder promedio ya que su fuerza es de ' + Fuerza . En el caso contrario 'Este pokemon' + nombre del pokemon + 'tiene menor poder promedio ya que su fuerza es de ' + Fuerza

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pokemon = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')

nombres_pokemon = pokemon['Name'].sort_values()
poderes = pokemon['HP']
poder_promedio = poderes.mean()


for nombre,poder in zip(nombres_pokemon,poderes):
    if poder >= poder_promedio:
        print("Este pokemon {} tiene mayor poder promedio ya que su fuerza es de {}".format(nombre,poder))
    elif poder < poder_promedio:
        print("Este pokemon {} tiene menor poder promedio ya que su fuerza es de {}".format(nombre,poder))
```

Haga 10 ejemplos entre palabras, números y operaciones matemáticas con las formas de óptimizar código mostrando los tiempos de ejecución con el comando %time o %timeit

```python
%time
print('Ejemplo 1:Este es un código poco eficiente')
double = []

for x in range(1,101):
    double.append((x ** 8)+5*(x**2))
print(double)

%time
print('Ejemplo 1:Este es un código super eficiente')
nums = list(range(1,101))
double = []
double = [(x ** 8)+5*(x**2) for x in nums]
print(double)
```

```python
%time
print('Ejemplo 2:Este es un código poco eficiente')
double = []

for x in range(0,101):
    if x==0:
        double.append(0)
    elif x==1:
        double.append(1)
    else:         
        double.append(double[x-1]+double[x-2])
print(double)

%time
print('Ejemplo 2:Este es un código muy eficiente con funciones')
from functools import reduce
fibonacci = lambda n: reduce(lambda double, _: double+[double[-1]+double[-2]],range(n-2), [0, 1])
print(fibonacci(101))                               
```

```python
%time
print('Ejemplo 3:Este es un código poco eficiente')
double = []
arreglo = []

for x in range(0,10):
    if x==0:
        double.append(1)
    elif x==1:
        double.append(1)    
    elif x==2:
        double.append(2)  
    else:         
        double.append(double[x-2]*double[x-1])
print(double)

%time
print('Ejemplo 3:Este es un código muy eficiente con funciones lambda')
#Factorial = lambda numero: reduce(lambda x, y: arreglo+[arreglo[-1], range(1, numero))
multip = lambda n: reduce(lambda arreglo, y: arreglo+[arreglo[-1]*arreglo[-2]],range(n-3), [1, 1, 2])
print(multip(10))
```

```python
%time
print('Ejemplo 4:Este es un código poco eficiente')
nombres=['Esta ','frase ','fue ','bien ','construida']
frase=''

for nombre in nombres:
    frase=frase+nombre
print(frase)


%time
print('Ejemplo 4:Este es un código muy eficiente')
nombres=['Esta ','frase ','fue ','bien ','construida']
frase=''

frase = lambda text: ''.join(text)
print(frase(nombres))
```
