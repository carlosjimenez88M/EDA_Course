```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Tarea 1
# Con base al dataframe de pokemon (El link se encuentra en la intro de la clase), va a crear un for loop con if cuando
# el poder promedio de los pokemones es superior se debe imprimir el mensaje
# 'Este pokemon' + nombre del pokemon + 'tiene mayor poder promedio ya que su fuerza es de ' + Fuerza .
# En el caso contrario 'Este pokemon' + nombre del pokemon + 'tiene menor poder promedio ya que su fuerza es de ' + Fuerza

pokemon = pd.read_csv('C:/Users/HARBEY/Downloads/archive/pokemon.csv')
```

```python
#Supongamos que hp es la variable que describe el poder de un Pokemon
name_pokemon = pokemon['name']
power = pokemon['hp']

print('El poder promedio de un Pokemon es', round(power.mean(),0))

for i,j in zip(name_pokemon, power):
    power_avg = power.mean()
    if j > power_avg:
        print('El pokemon {} tiene mayor poder promedio ya que su fuerza es de {}'.format(i,j))
    elif j < power_avg:
        print('El pokemon {} tiene menor poder promedio ya que su fuerza es de {}'.format(i,j))
```
