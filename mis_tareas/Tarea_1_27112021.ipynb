```python
Nombre_del_Pokemon = pokemon['Name']
Poder_del_Pokemon = pokemon['HP']
```

```python
pokemon[['Name','HP']]
```

```python
Poder_promedio = pokemon['HP'].median()
print('El poder promedio de los pokemones es de', Poder_promedio)

for i,j in zip(Nombre_del_Pokemon, Poder_del_Pokemon):
    if j > Poder_promedio:
        print('Este pokemon {} tiene mayor poder promedio ya que su fuerza es de {}'.format(i,j))
    elif j < Poder_promedio:
        print('Este pokemon {} tiene menor poder promedio ya que su fuerza es de {}'.format(i,j))
```
