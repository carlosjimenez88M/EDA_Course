## Función elif

Se usa cuando existe una segunda condición que deba ser valorada como un segundo paso

```python
nombres = base_estudiantes['Estudiantes']
Horas_de_estudio = base_estudiantes['Horas de estudio']

for i,hora_de_estudio in zip(nombres,Horas_de_estudio):
    horas_promedio_de_estudio = Horas_de_estudio.mean()
    if hora_de_estudio > horas_promedio_de_estudio:
        print("{} estudia normalmente {} horas a la semana, lo cual supera el promedio en {}".format(i,hora_de_estudio, round(abs(horas_promedio_de_estudio-hora_de_estudio)),0))
    elif hora_de_estudio < horas_promedio_de_estudio:
        print("{} estudia normalmente {} horas a la semana, lo cual esta por debajo de el promedio en {}".format(i,hora_de_estudio, round(abs(horas_promedio_de_estudio-hora_de_estudio)),0))
```

## Función While

Con esta función se ejecutan procesos hasta que se enceuentre el valor verdadero

```python
# Imprime i hasta que sea menor a 10
i = 1
while i < 10:
  print(i)
  i += 1
```

```python
## Esta se puede condicionar con la función break
i = 1
while i < 10:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
## COndicionando con la función continue
i = 0
while i < 10:
  i += 1
  if i == 3:
    continue
  print(i)
```
