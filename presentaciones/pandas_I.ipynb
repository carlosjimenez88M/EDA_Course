```python
## Librerias necesarias para esta clase
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pokemon = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')
pokemon = pokemon.drop(['#'],axis=1)
```

## Introducción  al manejo de bases de datos 

```python
# Con el comando head pueden verse los primero elementos de una base de datos 
pokemon.head()
```

```python
## Con la función tail se ven los últimos datos del frame
pokemon.tail()
```

```python
## Para ver la estructura de los datos se trabaja con .info
pokemon.info()
```

```python
## Para conocer las dimensiones de la base de dato se usa la extensión .shape
pokemon.shape
```

```python
## Para ver algunos datos estadísticos del set de datos se usa la función describe
pokemon.describe()
```

```python
## Para ver los datos o elementos de una sola columna se usa la extensión .values 
pokemon['Defense'][:3].values
```

```python
## Para ver las variables o columnas de una base de datos se usa la extensión .columns
pokemon.columns
```

```python
## Para organizar datos se usa la extensión sort_values

pokemon.sort_values(by='Total', ascending=False).head()
```

```python
pokemon.sort_values(by='Total', ascending=True).head()
```

```python
## Ordenando entre dos variables
pokemon.sort_values(by=['HP','Attack'], ascending=False).head()
```

```python
## Ordenando entre dos variables con una ascendente y otra descendente 
pokemon.sort_values(by=['HP','Attack'], ascending=[True,False]).head()
```

## Accediendo a elementos dentro de la base de datos 

```python
# Se accede usando el []
pokemon['Name']
```

```python
## Para acceder a más de uno es necesario usar [[]]
pokemon[['Name','Type 1']]
```

```python
## Se filtra de la siguiente manera 
# suponga que quiere filtrar los valores que estan encima de la mediana de los Ataque de cada uno de los pokemones

valor_a_filtrar = pokemon['Attack'].median()
pokemon[pokemon['Attack']>valor_a_filtrar]
```

```python
## Esto se puede visualizar de la siguiente manera
pokemon[pokemon['Attack']>valor_a_filtrar]['Attack'].hist()
plt.title('Valores de Attack por encima de la mediana')
```

```python
## En el caso que se quiera filtrar por los valores que esten por debajo de la mediana del ataque sería
pokemon[pokemon['Attack']<valor_a_filtrar]
```

```python
pokemon[pokemon['Attack']<valor_a_filtrar]['Attack'].hist()
plt.title('Valores de Attack por debajo de la mediana')
```

```python
pokemon['Attack'].hist()
plt.title('Valores de Attack sin filtros')
```

```python
## Suponga que ahora quiere filtrar por un tipo de variable cualitativa

pokemon.groupby(['Type 1']).size().reset_index(name='Total').sort_values(by='Total', ascending=False)
```

```python
pokemon[pokemon['Type 1']=='Water']
```

```python
## Para hacer filtro con varias condiciones
water = pokemon['Type 1']=='Water'
legendary = pokemon['Legendary']==True
pokemon[water & legendary]
```

```python
## Si se quiere filtrar por multiples condiciones en una sola variable
tipos = ['Water','Normal']
pokemon[pokemon['Type 1'].isin(tipos)]
```

```python
## Agregando nuevos columnas
pokemon.head()
pokemon['Ratio'] = pokemon['Defense']/pokemon['Attack']
pokemon.head()
```

```python
pokemon.Ratio.hist()
```

## Calculando algunas propiedades estadísticas

```python
## Calculando el promedio
pokemon['HP'].mean()
```

```python
## Calculando la mediana
pokemon['HP'].median()
```

```python
## Calculando la varianza
pokemon['HP'].var()
```

```python
## Calculando el máximo valor
pokemon['HP'].max()
```

```python
## Calculando el minimo
pokemon['HP'].min()
```

```python
## Calculando la desviación estandar
pokemon['HP'].std()
```

```python
## Calculando la suma de todos los valores 
pokemon['HP'].sum()
```

```python
## Calculando los quantiles
np.quantile(pokemon['HP'],q=(0.25,0.50,0.75))
```

```python
## Generando conteos
pokemon['Type 1'].value_counts()
```

```python
## Generando conteos
pokemon['Type 1'].value_counts(normalize=True)
```

```python
## Agrupando valores
pokemon.groupby('Type 1')['Total'].mean().reset_index(name='Avg_Total')
```

```python
## Agregando varias
pokemon.groupby('Type 1')['Total'].agg([min,max,np.mean,np.median]).reset_index()
```

## Visualización de datos 

```python
base_1 = pokemon.groupby('Type 1')['Total'].mean().reset_index(name='Total').sort_values(by='Total',ascending=True)
plt.barh(base_1['Type 1'],base_1['Total'])
```

```python
pokemon.plot('Type 1','Attack',kind='line')
pokemon.plot('Type 2','Attack',kind='line', color='red')
```

```python
pokemon.plot('HP','Attack',kind='scatter')
```

```python
pokemon[pokemon['Type 1']=='Water']['Total'].hist()
pokemon[pokemon['Type 1']=='Dragon']['Total'].hist()
plt.legend(["Water","Dragon"])
```
